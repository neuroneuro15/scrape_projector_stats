```python
from bs4 import BeautifulSoup
```

```python
with open('projectors-www.projectorcentral.com.html') as f:
    soup = BeautifulSoup(f.read(), 'lxml')
```

```python
table = soup.find('table', {'class': 'partList'})
table.text[:700]
```

```python
from datetime import datetime

projectors = []
for row in table.find_all('tr')[4::2]:
    strings = [el for el in row.stripped_strings]
    data = {}
    data['name'] = strings[0]
    data['release_date'] = datetime.strptime(strings[2], '%m/%d/%Y')
    data['price_dollars'] = int(strings[-1][1:].replace(',', ''))
    projectors.append(data)

projectors
    
        
```
